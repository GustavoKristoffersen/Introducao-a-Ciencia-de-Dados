#4.3 - Análise exploratória dos dados e mais gráficos

### Analisando alguns filmes

```python
df_filmes = pd.read_csv('./movies.csv')
df_filmes.columns = ['filmeId', 'titulo' , 'generos']
df_filmes
```

```python
#Notas do Toy Story
df.query('filmeId == 1')['nota']
```

```python
#Média
df.query('filmeId == 1')['nota'].mean()
```

### GroupBy

```python
medias_por_filme = df.groupby('filmeId')['nota'].mean()
medias_por_filme.head()
```

### Gráficos

```python
#histograma
medias_por_filme.plot(kind='hist')
```

```python
#Boxplot
sns.boxplot(medias_por_filme)
```

```python
#Medidas descritivas
medias_por_filme.describe()
```

```python
#histograma do Seaborn
sns.distplot(medias_por_filme)
```

```python
#Alterando a quantidade de separações
sns.distplot(medias_por_filme, bins=10)
```

```python
#Utilizando o Matplotlib
import matplotlib.pyplot as plt

plt.hist(medias_por_filme)
plt.title("Histograma das médias dos filmes")
```

```python
#Alterando o eixo do histograma com o matplotlib
sns.boxplot(y=medias_por_filme)
```

```python
plt.figure(figsize=(8,8))

sns.boxplot(y=medias_por_filme)
```
