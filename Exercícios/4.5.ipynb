```python
import pandas as pd
```

```python
df = pd.read_csv('./tmdb_5000_movies.csv')
df.head()
```

```python
df['original_language'].unique()
```

```python
#Descobrindo quantos valores de cada categoria existem
df['original_language'].value_counts()
```

```python
#Criando um dataframe com a categoria como índice
df['original_language'].value_counts().to_frame()
```

```python
#Criando um dataframe com a categoria como coluna
df['original_language'].value_counts().reset_index()
```

```python
language_counting = df['original_language'].value_counts().to_frame().reset_index()
language_counting.columns = ["original_language", "total"]

language_counting
```

```python
#Gráfico de barras
sns.barplot(x=language_counting['original_language'], y=language_counting['total'])
```
