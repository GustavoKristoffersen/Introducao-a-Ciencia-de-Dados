#3.14 - Exercício sobre APIs

```python
import requests
import json

api_key = '3fc009734313cac41e054afdaaa53fc9'
country = 'spain'
url = f'http://ws.audioscrobbler.com/2.0/?method=geo.getTopArtists&api_key={api_key}&country={country}&format=json'

response = requests.get(url)

if response.status_code == 200:
  data = json.loads(response.text)
  artists = data['topartists']['artist']
  top_artist = artists[0]

  top_artist = list(filter(lambda x: x['listeners'] > top_artist['listeners'], artists))
  
  print(top_artist)
else:
  print(response.content)
```
